```python
import pandas as pd
import os

DATA_DIR = 'data'
# 12個Excel檔案的路徑
file_names = ["chips24_1.csv","chips24_2.csv", "chips24_3.csv", "chips24_4.csv", "chips24_5.csv", "chips24_6.csv", "chips24_7.csv",
         "chips24_8.csv", "chips24_9.csv", "chips24_10.csv", "chips24_11.csv", "chips24_12.csv"]

combined_df_from_db = pd.DataFrame()
for file_name in file_names:
    data_path = os.path.join(DATA_DIR, file_name)
    df_from_db = pd.read_csv(data_path,na_values=[''])
    combined_df_from_db = pd.concat([combined_df_from_db, df_from_db], ignore_index=True)
combined_df_from_db['date'] = pd.to_datetime(combined_df_from_db['date'])
combined_df_from_db
# combined_df_from_db['investment_banks ']
```

```python
combined_df_from_db = combined_df_from_db.sort_values(by=['company_id', 'date'], ascending=[True, True])
# print(combined_df_from_db.dtypes)
# 將 以下欄位 NaN 值設為0
columns_to_fill = ['foreign_investors_day', 'investment_banks_days', 'dealer_days', 'all_institutional_investors_days']
#'directors_stock', 'financing_balance','margin_balance','margin_balance_ratio', 'financing_balance_ratio','sale_margin_purchase_ratio', 'directors_stock_ratio','sale_margin_purchase_ratio', 
# combined_df_from_db[columns_to_fill] = pd.to_numeric(combined_df_from_db[columns_to_fill], errors='coerce').fillna(0)
combined_df_from_db[columns_to_fill] = combined_df_from_db[columns_to_fill].fillna(0)
# for column in columns_to_fill:
    # # 先處理千分位表示的數字，將逗號移除
    # combined_df_from_db[column] = combined_df_from_db[column].str.replace(',', '')

    # 使用 pd.to_numeric 將欄位轉換為數值，並將無法轉換的值設為 NaN
    # combined_df_from_db[column] = pd.to_numeric(combined_df_from_db[column], errors='coerce').fillna(0)
# combined_df_from_db["foreign_investors"].apply(pd.to_numeric, errors='coerce').fillna(0)
# combined_df_from_db["foreign_investors"]
#TODO 轉換千分位符號
```

```python

# 將 '200_400_ratio', '400_600_ratio', '600_800_ratio', '800_1000_ratio', '1000up_ratio' 這幾欄的 NaN 值用前一列的值填補，如果前一列也是 NaN 則繼續往前找有值的值填補
combined_df_from_db[['200_400_ratio', '400_600_ratio', '600_800_ratio', '800_1000_ratio', '1000up_ratio']] = combined_df_from_db[['200_400_ratio', '400_600_ratio', '600_800_ratio', '800_1000_ratio', '1000up_ratio']].fillna(method='ffill')
combined_df_from_db
```
